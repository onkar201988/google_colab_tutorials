```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
np.random.seed(0)
points = 500
x = np.linspace(-3, 3, points) # 500 points between -3 to 3
y = np.sin(x) + np.random.uniform(-0.5, 0.5, points)

plt.scatter(x, y)
```

```python
model = Sequential()
model.add(Dense(50, input_dim=1, activation='sigmoid')) # X is the input and Y is output, hence there is 1 input
model.add(Dense(30, activation='sigmoid'))
model.add(Dense(1))
model.compile(Adam(learning_rate=0.01), loss='mse') # mean square error
model.fit(x=x, y=y, epochs=50)
```

```python
prediction = model.predict(x)
plt.scatter(x, y)
plt.plot(x, prediction, 'ro')
```
